```python
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("medicare_data.csv")
df
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='Rndrng_Prvdr_CCN', data=df, order=df['Rndrng_Prvdr_CCN'].value_counts().index)
plt.title('Rendering Provider CCN Count')
plt.xlabel('Count')
plt.ylabel('Rendering Provider CCN')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Tot_Dschrgs'], kde=True)
plt.title('Total Discharges Distribution')
plt.xlabel('Total Discharges')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Avg_Mdcr_Pymt_Amt'], kde=True)
plt.title('Average Medicare Payment Amount Distribution')
plt.xlabel('Average Medicare Payment Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(y='Avg_Mdcr_Pymt_Amt', data=df)
plt.title('Boxplot of Average Medicare Payment Amount')
plt.xlabel('Average Medicare Payment Amount')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Avg_Submtd_Cvrd_Chrg', y='Avg_Mdcr_Pymt_Amt', data=df)
plt.title('Scatterplot of Submitted Covered Charge vs. Medicare Payment Amount')
plt.xlabel('Average Submitted Covered Charge')
plt.ylabel('Average Medicare Payment Amount')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Avg_Tot_Pymt_Amt'], kde=True)
plt.title('Average Total Payment Amount Distribution')
plt.xlabel('Average Total Payment Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Avg_Submtd_Cvrd_Chrg'], kde=True)
plt.title('Average Submitted Covered Charge Distribution')
plt.xlabel('Average Submitted Covered Charge')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Tot_Dschrgs'], kde=True)
plt.title('Total Discharges Distribution')
plt.xlabel('Total Discharges')
plt.ylabel('Frequency')
plt.show()
```

```python
print(df.info())
print(df.describe())

# Noma'lum qiymatlarni tekshirish
print(df.isnull().sum())

# 1. Boxplot - Avg_Tot_Pymt_Amt bo'yicha
plt.figure(figsize=(10, 6))
sns.boxplot(y='Avg_Tot_Pymt_Amt', data=df)
plt.title('Boxplot of Average Total Payment Amount')
plt.xlabel('Average Total Payment Amount')
plt.show()

# 2. Violinplot - Avg_Tot_Pymt_Amt bo'yicha
plt.figure(figsize=(10, 6))
sns.violinplot(y='Avg_Tot_Pymt_Amt', data=df)
plt.title('Violinplot of Average Total Payment Amount')
plt.xlabel('Average Total Payment Amount')
plt.show()

# 3. Scatterplot - Avg_Tot_Pymt_Amt va Avg_Submtd_Cvrd_Chrg o'rtasida
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Avg_Submtd_Cvrd_Chrg', y='Avg_Tot_Pymt_Amt', data=df)
plt.title('Scatterplot of Submitted Covered Charge vs. Total Payment Amount')
plt.xlabel('Average Submitted Covered Charge')
plt.ylabel('Average Total Payment Amount')
plt.show()

# 4. Pairplot - Ba'zi ustunlar o'rtasidagi munosabatni ko'rsatish
sns.pairplot(df[['Tot_Dschrgs', 'Avg_Submtd_Cvrd_Chrg', 'Avg_Tot_Pymt_Amt', 'Avg_Mdcr_Pymt_Amt']])
plt.show()
```
