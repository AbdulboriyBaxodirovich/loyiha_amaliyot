```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
df = df.reindex(range(1, len(df) + 1))
print(df)
```

## Math_Score uchun grafik chizamiz.

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.plot(df.index, df["Math_Score"])
ax.plot(df.index, df["Writing_Score"])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(df.index, df["Math_Score"], marker = "o")
ax.plot(df.index, df["Writing_Score"], marker = "o")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(df["Math_Score"], df["Reading_Score"])
plt.show()
```

```python
fig, ax = plt.subplots()
math_sc = df[df["Math_Score"] == 75]
reading_sc = df[df["Reading_Score"] == 88]

std1 = math_sc["Reading_Score"].std()
std2 = reading_sc["Math_Score"].std()

ax.bar("O'rtacha reading", math_sc['Reading_Score'].mean(), yerr = std1)
ax.bar("O'rtacha math", reading_sc['Math_Score'].mean(), yerr = std2)
```

```python
fig, ax = plt.subplots(2, 1)
math_sc = df[df["Math_Score"] == 75]
reading_sc = df[df["Reading_Score"] == 88]

std1 = math_sc["Reading_Score"].std()
std2 = reading_sc["Math_Score"].std()

ax[0].bar("O'rtacha reading", math_sc['Reading_Score'].mean(), yerr = std1)
ax[0].bar("O'rtacha math", reading_sc['Math_Score'].mean(), yerr = std2)

ax[1].scatter(df["Math_Score"], df["Reading_Score"])
```

```python
fig, ax = plt.subplots()
plt.style.use("ggplot")
ax.plot(df.index, df["Math_Score"])
ax.plot(df.index, df["Writing_Score"])
plt.show()
```
