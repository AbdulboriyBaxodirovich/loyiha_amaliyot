## Bu holatda NaN ning o'rniga tahminan medianani qo'yib ketsak bo'ladi. Chunki bo'sh qatorlar deyarli 0.5 % ni tashkil qiladi. Bu holatda esa xatolik juda kichik va aniqlik 1ga yaqin bo'ladi.

```python
df["Field_Goal_Prob"].median()
```

```python
df["Field_Goal_Prob"] = df["Field_Goal_Prob"].fillna(value=df["Field_Goal_Prob"].median())
df
```
