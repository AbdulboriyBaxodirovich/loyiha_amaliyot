```python
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("insurance.csv")
df
```

```python
df = pd.get_dummies(data=df, columns=['sex'], drop_first=True, dtype=int)
df
```

```python
df = pd.get_dummies(data=df, columns=['smoker'], drop_first=True, dtype=int)
df
```

```python
df.describe()
```

```python
df["bmi"].hist()
```

```python
avv = df["bmi"].quantile(0.25)
key = df["bmi"].quantile(0.75)
std = df["bmi"].std()

filtered_bmi = df[(df["bmi"] > avv - 3 * std) & (df["bmi"] < key + 3 * std)]["bmi"]

plt.hist(filtered_bmi, bins=10, edgecolor='black')
plt.xlabel('BMI')
plt.show()
```

```python
avv = df["age"].quantile(0.25)
key = df["age"].quantile(0.75)
std = df["age"].std()

filtered_bmi = df[(df["age"] > avv - 3 * std) & (df["age"] < key + 3 * std)]["age"]

plt.hist(filtered_bmi, bins=10, edgecolor='black')
plt.xlabel('age')
plt.show()
```
