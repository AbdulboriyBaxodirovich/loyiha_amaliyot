```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("processed_data.csv")
df.head()
```

```python
df["Academic Competitions"].value_counts()
```

```python
sns.set_style("whitegrid")

sns.catplot( x= "Academic Competitions", y = "Communication ability", data = df, hue = "Gender", kind = "point")

plt.xticks(rotation = 90)
plt.show()
```

```python
sns.set_style("whitegrid")

sns.countplot( x = "Summer School", hue = "Summer School", data = df)

plt.xticks(rotation = 90)
plt.show()
```

```python
sns.set_style("whitegrid")

plot = sns.catplot(x="Academic Competitions", y="Communication ability", hue="Gender", data=df, kind="point")

plot.figure.suptitle('TITLE') 

plt.xticks(rotation = 90)
plt.show()
```

```python
sns.set_style("whitegrid")

plot = sns.catplot(x="Academic Competitions", y="Communication ability", hue="Gender", data=df, kind="point")

plot.figure.suptitle('TITLE', y = 1.03) 

plt.xticks(rotation = 90)
plt.show()
```
