```python
import pandas as pd
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
plt.show()
```

```python
seatter_weather = pd.read_csv("seattle_weather.csv")
seatter_weather.head()
```

## Seatter shahrida kuzatilgan havo haroratining grafigi.

```python
fig, ax = plt.subplots()
ax.plot(seatter_weather["DATE"].head(12), seatter_weather["MLY-TAVG-NORMAL"].head(12))
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(seatter_weather["DATE"].head(12), seatter_weather["MLY-TAVG-NORMAL"].head(12), marker = "o")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(seatter_weather["DATE"].head(12), seatter_weather["MLY-TAVG-NORMAL"].head(12), marker = "o", linestyle = "--")
plt.show()
```

## Seatter shahridagi MLY-TMIN-PRBOCC-LSTH020 ning grafigi.

```python
fig, ax = plt.subplots()
ax.plot(seatter_weather["DATE"].head(12), seatter_weather["MLY-TMIN-PRBOCC-LSTH020"].head(12), marker = "o", linestyle = "--")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(seatter_weather.index, seatter_weather["MLY-TAVG-NORMAL"])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(seatter_weather.index.sort_values(), seatter_weather["MLY-TMIN-PRBOCC-LSTH020"].sort_values())
plt.show()
```
