```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv("processed_data.csv")
df.head()
```

```python
sns.stripplot(data = df, y = "Creativity ability", x = "Communication ability", hue="Gender")
```

```python
sns.swarmplot(data = df, y = "Creativity ability", x = "Communication ability", hue="Gender")
```

```python
sns.boxplot(data = df, y = "Creativity ability", x = "Communication ability", hue="Gender")
```

```python
sns.violinplot(data = df, y = "Creativity ability", x = "Communication ability", hue="Gender")
```

```python
sns.boxenplot(data = df, y = "Creativity ability", x = "Communication ability", hue="Gender")
```

```python
sns.barplot(data = df, y = "Creativity ability", x = "Communication ability", hue="Gender")
plt.xticks(rotation=90, ha = "right")
```

```python
bins = pd.cut(df["Communication ability"], bins=5, labels=False)

df["Comm Ability Group"] = bins

sns.violinplot(data=df, y="Creativity ability", x="Comm Ability Group", hue="Gender")
plt.show()
```

```python
sns.regplot(data=df, x = "Communication ability", y = "Creativity ability", marker="+")
```

```python
sns.residplot(data=df, x = "Communication ability", y = "Creativity ability")
```

```python
sns.regplot(data=df, x = "Communication ability", y = "Creativity ability", x_jitter=.1, order=2)
```

```python

x_jitter_value = df["Communication ability"].mean() 

sns.regplot(data=df, x="Communication ability", y="Creativity ability", x_jitter=x_jitter_value, order=2)
```

```python
sns.regplot(data=df, x = "Communication ability", y = "Creativity ability", x_bins=10, order=2)
```
