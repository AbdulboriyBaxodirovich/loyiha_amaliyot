```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
medals = pd.read_csv("medals_by_country_2016.csv")
medals.head()
```

```python
arr1 = np.array(medals["Bronze"])
arr2 = np.array(medals["Gold"])
sns.scatterplot(x = arr1, y = arr2)
plt.show()
```

## Medallar taqsimotini tuzing..

```python
md = np.array(medals["Silver"])
sns.countplot(x = md)
plt.show()
```

```python
sns.countplot(x = "Gold", data = medals)
plt.show()
```

```python
sns.scatterplot(x = "Gold", y = "Silver", data = medals)
plt.show()
```

## Har bir davlat nechtadan Kumush va Oltin medal olganini ko'rsating.

```python
sns.scatterplot(x = "Gold", y = "Silver", data = medals, hue="Country")
plt.show()
```

```python
hue_colors = {
    "United States": "blue",
    "Germany": "green",
    "Great Britain": "red",
    "Russia": "purple",
    "China": "orange",
    "France": "cyan",
    "Australia": "magenta",
    "Italy": "yellow",
    "Canada": "black",
    "Japan": "brown"
}
sns.scatterplot(x = "Gold", y = "Silver", data = medals, hue="Country", palette = hue_colors)
plt.show()
```

```python
sns.barplot(x ="Country", y = "Gold", data = medals)
plt.xticks(rotation=45)
plt.show()
```
